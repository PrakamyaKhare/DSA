```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def hailstone_number(N) :
    data = []
    data.append(N)
    
    while N != 1 :
        if N%2 == 0 :
            N = N/2
            data.append(N)
        else :
            N = 3*N + 1
        
    return data
```

```python
def plot_data(X=None,y=None) :
    plt.plot(y)
    plt.show()
```

```python
y = hailstone_number(7)
plot_data(y=y)
```

```python
plot_data(y=hailstone_number(90))
```
